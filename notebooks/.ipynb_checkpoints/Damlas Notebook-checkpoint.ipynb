#### **Metadata**

This comprehensive set of fields will guide your analysis, helping you unravel the intricacies of client behavior and preferences.

client_id: Every client’s unique ID.

variation: Indicates if a client was part of the experiment.

visitor_id: A unique ID for each client-device combination.

visit_id: A unique ID for each web visit/session.

process_step: Marks each step in the digital process.

date_time: Timestamp of each web activity.

clnt_tenure_yr: Represents how long the client has been with Vanguard, measured in years.

clnt_tenure_mnth: Further breaks down the client’s tenure with Vanguard in months.

clnt_age: Indicates the age of the client.

gendr: Specifies the client’s gender.

num_accts: Denotes the number of accounts the client holds with Vanguard.

bal: Gives the total balance spread across all accounts for a particular client.

calls_6_mnth: Records the number of times the client reached out over a call in the past six months.

logons_6_mnth: Reflects the frequency with which the client logged onto Vanguard’s platform over the last six months.

```python
df_web_data = pd.concat([df_pt1, df_pt2], axis=0, ignore_index=True)
df_web_data
```

```python
df_web_data.dtypes
```

```python
df_web_data['date_time'] = pd.to_datetime(df_web_data['date_time'])

# Display the DataFrame with updated data types
print("\nDataFrame with updated data types:")
print(df_web_data.dtypes)
```

```python
unique_visits_per_client = df_web_data.groupby('client_id')['visit_id'].nunique().reset_index(name='unique_visits')

# Sort by 'unique_visits' descending
unique_visits_per_client_sorted = unique_visits_per_client.sort_values(by='unique_visits', ascending=False)

df_sorted_results = unique_visits_per_client_sorted

# Display the new DataFrame with sorted results
df_sorted_results
```

```python

#With this code,we reached a db that shows us; which unique clients had
#more than 1 visit(27.861)from 120.000 client
#more than 2 visit: 2926 client
#more than 3 visi: 2229 client
#more than 4 visit: 904 client
#more than 5 visit

df =df_sorted_results[df_sorted_results["unique_visits"]>0]
```

```python
df
```

```python
#Counting occurrences of each unique_visits value
visit_counts = df['unique_visits'].value_counts().sort_index()

# Plotting the bar plot
plt.figure(figsize=(10, 6))
bars = visit_counts.plot(kind='bar', color='skyblue')

# Adding annotations
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.2, int(yval), ha='center', va='bottom')

plt.xlabel('Unique Visits')
plt.ylabel('Number of Clients')
plt.title('Number of Clients by Unique Visits')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# 4)How many times this client go to confirm prosess 
# Generating a crosstab for 'client_id' and 'process_step'
df_web_data_crosstab = pd.crosstab(df_web_data['visit_id'], df_web_data['process_step'])
df_web_data_crosstab.sort_values(by='confirm', ascending = False)
```

```python
#If there are this steps for every visit_id=  a start-step-1-step2-step3-confirm 
time start<timestep1< time step2<time step3< confirm in visit id
#than we take them as a 1 confirm.
#we assume that every visit ID can have only 1 confirm.
#We calculate for every visit ID how much time customer spend?
#We assume that if a customer in this order step1-step2-step3 and confirm go back and forward, than this costumer has confusion.
#We assume that if the customers do a step more than 1 in a row: There may be an error.
#If the order of the process kronogically wrong for example a customer starsts his prosesss with confirm or step2 :there is an arror.
#We still take this custumers if they this all steps later cronologically and finish with confirm: as a confirm.
#Not confirmed customers are: if there is a missing step,
#                             if there is no cronologically ordered start-step1-2-3-confirm in their prosess  of the confirmation 
# We can take the visitor id, and if this visitor id is test or control group in data frame.                             
```
